```python
import os
cwd = os.getcwd()
cwd

from visualise_tools import *

EXPERIMENTS = [
    (f'output/img/tracknoobstacles-trackall', "empty-field-1000-cells-5000-timeunits", 1000),
    (f'output/img/trackobstacles5by5-trackall', "5x5-field-1000-cells-5000-timeunits", 1000),
    (f'output/img/trackobstacles-trackall', "10x10-field-1000-cells-5000-timeunits", 1000),
    (f'output/img/trackobstacles15by15-trackall', "15x15-field-1000-cells-5000-timeunits", 1000),
    (f'output/img/trackobstacles15by15-track100', "15x15-field-100-cells-5000-timeunits", 100),
    (f'output/img/tracknoobstacles-track100', "empty-field-100-cells-5000-timeunits", 100)        
]
```

```python
for file, EXPERIMENT, cell_nr in EXPERIMENTS:
    n=0
    file_paths = [f'{file}{n}/position{i}.txt' for i in range(1,cell_nr+2)]  # Replace with actual file paths
    all_positions = load_cell_data(file_paths)
    
    # Centroid Movement
    plot_centroid_movement(all_positions)
    # Show the plot
    print(EXPERIMENT, f"../../figures/{EXPERIMENT}-migration.png")
    plt.savefig(f"../../figures/{EXPERIMENT}-migration.png", bbox_inches='tight', pad_inches = 0)
    
    # Directionality
    plot_final_positions_with_direction(all_positions, 'output/img/empty/trackposition-t0.png')
    plt.savefig(f"../../figures/{EXPERIMENT}-directionality.png", bbox_inches='tight', pad_inches = 0)
    
    # Speed 
    plot_speeds(all_positions)
    ax = plt.gca()
    ax.set_ylim([0, 0.5])
    plt.savefig(f"../../figures/{EXPERIMENT}-speed.png", bbox_inches='tight', pad_inches = 0)
```
