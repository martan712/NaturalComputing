```python
from itertools import takewhile
from PIL import Image

def show_path(path, im, dark=False, custom = None, UNTIL = None):
    main_path = "output/img"
        
    def filter_tuples(lst):
        if not lst:
            return []

        filtered = [lst[0]]  # Start with the first element
        for i in range(1, len(lst)):
            x_prev, y_prev = filtered[-1]
            x_curr, y_curr = lst[i]

            if abs(x_curr - x_prev) > 0.8 or abs(y_curr - y_prev) > 0.8:
                break  # Stop processing if the condition is met

            filtered.append(lst[i])  # Otherwise, add to result

        return filtered
                        
    def read_file_to_tuples(file_path):
        with open(f"{file_path}", 'r') as file:
            lst = [tuple(map(float, line.strip().split(','))) for line in file if line.strip()]
            filtered_lst = filter_tuples(lst)
            return [(int(x[0]), int(x[1])) for x in filtered_lst]

    file_path = f"{main_path}/{path}/position.txt"
    coordinates = read_file_to_tuples(file_path)

    if UNTIL == None:
        UNTIL = len(coordinates)

    pixelMap = im.load()

    img = Image.new( im.mode, im.size)
    pixelsNew = img.load()
    for i in range(img.size[0]):
        for j in range(img.size[1]):
            pixelsNew[i,j] = pixelMap[i,j]
        
    for a, (i,j) in enumerate(coordinates[:UNTIL]):
        
        if custom: 
            for x in range(-2,2):
                for y in range(-2,2):
                    pixelsNew[i*2+x,j*2+y] = custom
                    if x ==y and x==0:
                        pixelsNew[i*2+x,j*2+y] = custom
                
        else:
            # lightmode (busy)
            if not dark:
                for x in range(-2,2):
                    for y in range(-2,2):
                        pixelsNew[i*2+x,j*2+y] = (255,int(255*((a+1)/UNTIL)),0,255)
                        if x ==y and x==0:
                            pixelsNew[i*2+x,j*2+y] = (255,int(255*((a+1)/UNTIL)),0,255)

            # darkmode (calm)
            if dark:
                for x in range(-2,2):
                    for y in range(-2,2):
                        pixelsNew[i*2+x,j*2+y] = (0,int(125*(1-(a+1)/UNTIL)),255,255)#(0,0,int(255*((a+1)/UNTIL)),255)
                        if x ==y and x==0:
                            pixelsNew[i*2+x,j*2+y] = (0,int(125*(1-(a+1)/UNTIL)),255,255)

    return img
```
