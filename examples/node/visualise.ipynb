```python
def read_file_to_tuples(file_path):
    with open(file_path, 'r') as file:
        lst = [tuple(map(float, line.strip().split(','))) for line in file if line.strip()]
        return [(int(x[0]), int(x[1])) for x in lst]

# Example usage
file_path = 'position.txt'  # Replace with your file path
coordinates = read_file_to_tuples(file_path)
print(coordinates)
```

```python
def show_path(path, dark=False):
    main_path = "output/img"
    def read_file_to_tuples(file_path):
        with open(f"{file_path}", 'r') as file:
            lst = [tuple(map(float, line.strip().split(','))) for line in file if line.strip()]
            return [(int(x[0]), int(x[1])) for x in lst]


    file_path = f"{main_path}/{path}/position.txt"
    coordinates = read_file_to_tuples(file_path)

    UNTIL = 3400

    from PIL import Image
    im = Image.open(f"{main_path}/{path}/trackposition-t{UNTIL}.png")
    pixelMap = im.load()

    img = Image.new( im.mode, im.size)
    pixelsNew = img.load()
    for i in range(img.size[0]):
        for j in range(img.size[1]):
            pixelsNew[i,j] = pixelMap[i,j]
        
    for a, (i,j) in enumerate(coordinates[:UNTIL]):
        # lightmode (busy)
        if not dark:
            for x in range(-2,2):
                for y in range(-2,2):
                    pixelsNew[i*2+x,j*2+y] = (255,int(255*((a+1)/UNTIL)),255,255)
                    if x ==y and x==0:
                        pixelsNew[i*2+x,j*2+y] = (0,0,0,255)

        # darkmode (calm)
        if dark:
            for x in range(-2,2):
                for y in range(-2,2):
                    pixelsNew[i*2+x,j*2+y] = (0,int(125*(1-(a+1)/UNTIL)),255,255)#(0,0,int(255*((a+1)/UNTIL)),255)
                    if x ==y and x==0:
                        pixelsNew[i*2+x,j*2+y] = (0,int(125*(1-(a+1)/UNTIL)),255,255)

    img.show()
    img.save(f"{main_path}/{path}/trace.png")
    
```

```python
show_path("trackposition-long-empty", dark=True)
```

```python
show_path("trackobstacles", dark=True)
```
